```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


file_path = '/content/jeet.csv'
data = pd.read_csv(file_path)


filtered_data = data.dropna(subset=["Life expectancy", "Health expenditure"])


filtered_data["Scaled Health Expenditure"] = filtered_data["Health expenditure"] / 1e8

x = np.arange(len(filtered_data["Entity"]))
width = 0.35

fig, ax = plt.subplots(figsize=(14, 7))

ax.bar(x - width/2, filtered_data["Life expectancy"], width, label="Life Expectancy", color="skyblue")

ax.bar(x + width/2, filtered_data["Scaled Health Expenditure"], width, label="Health Expenditure (scaled)", color="orange")

ax.set_xlabel("Entity")
ax.set_ylabel("Values")
ax.set_title("Life Expectancy and Health Expenditure by Entity")
ax.set_xticks(x)
ax.set_xticklabels(filtered_data["Entity"], rotation=90)
ax.legend()

plt.tight_layout()
plt.show()
```
